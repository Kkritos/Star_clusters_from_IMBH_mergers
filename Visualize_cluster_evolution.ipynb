```python
N = 300

xi_e, zeta = 0.0074, 0.0926
#xi_e, zeta = 0.0450, 0.0725

f_s = 0.50

vesc_crit = 200

fig, ax = plt.subplots(figsize=(6, 4))

M_cl0 = np.logspace(3.8, 6.6, N)
r_h0 = np.logspace(-1.6, 1, N)

gridX, gridY = np.meshgrid(M_cl0, r_h0)

gridZ_flatten = []
for (x, y) in zip(gridX.flatten(), gridY.flatten()):
    if 2 * np.sqrt(kappa * G_grav * x / y) > vesc_crit or \
    w_cc * 0.138 / m_star0 / psi / lnL * np.sqrt(x * y**3 / G_grav) < 2:
        gridZ_flatten.append(np.nan)
    else:
        gridZ_flatten.append((black_hole_mass(t_Hub, N_star0=x / m_star0, r_h0=y, xi_e=xi_e, zeta=zeta, f_s=f_s)))

gridZ = np.reshape(gridZ_flatten, (N, N))

levels = [100, 200, 400, 800, 1600, 3200, 6400]
plt.contourf(M_cl0, r_h0, gridZ, levels, cmap='gnuplot')
cbar = plt.colorbar()
cbar.set_label('$M_{\\rm BH}(t=\\tau_{{\\rm Hub}})/M_\odot$', size=15)


plt.tick_params(which='both', direction='in', labelsize=15, top=True, right=True)

plt.loglog()
plt.xlim(M_cl0[0], M_cl0[-1])
plt.ylim(r_h0[0], r_h0[-1])

plt.xlabel('$M_{\\rm cl,0}/M_\odot$', size=20)
plt.ylabel('$r_{\\rm h,0}/{\\rm pc}$', size=20)

m = np.logspace(3, 7)
plt.plot(m, (G_grav / m * (t_Hub * m_star0 * psi * lnL / 0.138 / w_BH)**2)**(1/3), lw=1, color='royalblue', 
         ls='--')
plt.plot(m, (G_grav / m * (2 * m_star0 * psi * lnL / 0.138 / w_cc)**2)**(1/3), lw=1, color='darkorange', 
         ls='--')
plt.plot(m, kappa * G_grav * m / (vesc_crit/2)**2, lw=1, color='forestgreen', 
         ls='--')

plt.text(4.5e5, 0.055, '$v_{\\rm esc,0}>200\,\\rm km\, s^{-1}$', color='forestgreen', size=15, rotation=40)
plt.text(4e5, 3.00, '$\\tau_{_{\\rm BH}}>\\tau_{\\rm Hub}$', color='royalblue', size=15, rotation=-17)
plt.text(7e3, 3e-2, '$\\tau_{\\rm cc}<2\,\\rm Myr$', color='darkorange', size=15, rotation=-17)

ax.add_artist(AnchoredText("$\\rm Isolated$", loc=4, prop=dict(size=15), frameon=False))

ax.add_artist(AnchoredText("$f_{\\rm s}=50\%$", loc=2, prop=dict(size=15), frameon=True))

plt.tight_layout()
plt.savefig('Mcl0_rh0__MBH_tHub_Isolated_fs050.pdf')
```
