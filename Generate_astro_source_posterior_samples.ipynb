```python
def generate_astro_source_posterior_samples(PE_source_posterior_samples, log_prior_astro, 
                                            N_samples=1000, seed=5673435):
    """
    Generate posterior source samples assuming an astrophysical prior.
    
    Inputs:
    @in PE_source_posterior_samples: source posterior samples with an arbitrary prior assuming in the PE; 
                                     requires keys: "mass_1_source", "mass_2_source", "merger_time" (in Gyr), and
                                     "log_prior_source"
    @in log_prior_astro: astrophysical log-prior (function object that takes m1, m2, and zm as three arguments)
    @in N_samples: number of samples to generate
    @in seed: pseudo-random number seed
    
    Outputs:
    @out astro_source_posterior_samples: source posterior samples with the astrophysical prior with keys:
                                         "mass_1_source", "mass_2_source", and "merger_time"
    """
    
    # initialize pseudo-number generator:
    np.random.seed(seed)
    
    # number of source PE samples:
    N_PE_source_posterior_samples = PE_source_posterior_samples['log_prior_source'].size
    
    # compute and normalize astrophysical source prior:
    log_prior_astrophysical = np.array([log_prior_astro(PE_source_posterior_samples['mass_1_source'][i],
                                                        PE_source_posterior_samples['mass_2_source'][i],
                                                        PE_source_posterior_samples['redshift'][i]
                                                       ) for i in range(N_PE_source_posterior_samples)])
    log_prior_astrophysical = np.concatenate(log_prior_astrophysical - np.max(log_prior_astrophysical))
    
    # get and normalized PE source prior:
    log_prior_PE = PE_source_posterior_samples['log_prior_source']
    log_prior_PE = log_prior_PE - np.max(log_prior_PE)
    
    # normalized source prior density:
    p = np.exp(log_prior_astrophysical - log_prior_PE); p = p / np.sum(p)
    
    # generate astrophysical posterior source sample indices (sample with repetition and weights p):
    indices = np.random.choice(np.arange(N_PE_source_posterior_samples), replace=True, size=N_samples, p=p)
    
    # get astrophysical posterior source samples:
    astro_source_posterior_samples = {
        'mass_1_source': PE_source_posterior_samples['mass_1_source'][indices].values, 
        'mass_2_source': PE_source_posterior_samples['mass_2_source'][indices].values, 
        'merger_time': PE_source_posterior_samples['merger_time'][indices].values
    }
    
    return astro_source_posterior_samples
```
