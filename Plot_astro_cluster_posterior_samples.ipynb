```python
def HELLINGER_DISTANCE(d1, d2, xmin, xmax):
    
    p1 = gaussian_kde(d1)
    p2 = gaussian_kde(d2)
    
    dH = np.sqrt(quad(lambda x: (np.sqrt(p1(x)) - np.sqrt(p2(x)))**2, xmin, xmax)[0] / 2)
    
    return dH
```
